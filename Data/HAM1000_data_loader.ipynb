```python
import kagglehub
import shutil
from pathlib import Path

# Download dataset (cached by kagglehub)
src_path = Path(kagglehub.dataset_download("kmader/skin-cancer-mnist-ham10000"))

# Desired destination inside your project
dst_path = Path("../Data/HAM10000")

# Create destination if it doesn't exist
dst_path.parent.mkdir(parents=True, exist_ok=True)

# Move if not already moved
if not dst_path.exists():
    shutil.move(src_path, dst_path)

print("Dataset available at:", dst_path.resolve())
```
