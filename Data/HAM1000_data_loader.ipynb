```python
import kagglehub
import shutil
from pathlib import Path

# Download dataset (cached by kagglehub)
src_path = Path(kagglehub.dataset_download("kmader/skin-cancer-mnist-ham10000"))

# Desired destination inside your project
dst_path = Path("../Data/HAM10000")

# Create destination if it doesn't exist
dst_path.parent.mkdir(parents=True, exist_ok=True)

# Copy contents (not the version folder itself)
for item in src_path.iterdir():
    target = dst_path / item.name
    if target.exists():
        continue
    if item.is_dir():
        shutil.copytree(item, target)
    else:
        shutil.copy2(item, target)

print("HAM10000 dataset available at:", dst_path.resolve())
```
