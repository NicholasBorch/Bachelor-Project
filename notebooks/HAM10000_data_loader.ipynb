```python
import kagglehub
import shutil
from pathlib import Path
```

```python
# Project path for dataset
PROJECT_ROOT = Path("..").resolve()
DATA_ROOT = PROJECT_ROOT / "data"
RAW_ROOT = DATA_ROOT / "raw"

# Download dataset (cached by kagglehub)
src_path = Path(kagglehub.dataset_download("kmader/skin-cancer-mnist-ham10000"))

# Destination inside repo
dst_path = RAW_ROOT / "HAM10000"
dst_path.mkdir(parents=True, exist_ok=True)

# Copy contents
for item in src_path.iterdir():
    target = dst_path / item.name
    if target.exists():
        continue
    if item.is_dir():
        shutil.copytree(item, target)
    else:
        shutil.copy2(item, target)

print("HAM10000 dataset available at:", dst_path.resolve())
```
